```python
from sklearn.datasets import make_blobs
```

```python
raw_data = make_blobs(
    n_samples = 300, 
    n_features = 2, 
    centers = 5, 
    cluster_std = 2.8
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
plt.scatter(raw_data[0][:,0], raw_data[0][:,1], c=raw_data[1]);
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4)
model.fit(raw_data[0])
```

```python
modelPP = KMeans(n_clusters=4, init='k-means++')
modelPP.fit(raw_data[0])
```

```python
model.labels_
```

```python
modelPP.labels_
```

```python
model.cluster_centers_
```

```python
modelPP.cluster_centers_
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True,figsize=(15,6))
ax1.set_title(' ')
ax1.scatter(raw_data[0][:,0], raw_data[0][:,1],c=model.labels_)
ax1.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=200, c='red')
ax2.set_title('  PP')
ax2.scatter(raw_data[0][:,0], raw_data[0][:,1],c=modelPP.labels_)
ax2.scatter(modelPP.cluster_centers_[:, 0], modelPP.cluster_centers_[:, 1], s=200, c='red')
ax3.set_title(' ')
ax3.scatter(raw_data[0][:,0], raw_data[0][:,1],c=raw_data[1]);
```
